```python
import pandas as pd
import matplotlib.pyplot as plt 
import matplotlib.font_manager as font_manager
import pickle
import numpy as np
import os
import regex as re
import mandar_sql as rodar 

from scipy.interpolate import interp1d
from scipy.stats import linregress

from numpy import diff
from sklearn.preprocessing import PolynomialFeatures

from lmfit import Model
from sklearn.metrics import mean_squared_error
```

```python
#Função construída por James Moraes de Almeida
def XANES_E0(df):
    x = df.iloc[:, 0].values
    y =  df.iloc[:, 1].values
    dydx = diff(y)/diff(x)
    
    #plt.plot(x, y, label='Input Data', color='b') #DEBUG
    x = df.iloc[0:-1, 0].values
    #plt.plot(x,dydx) #DEBUG
    
    #Get the largest derivative
    ##There might be problems if a larger derivative is found beyond the first maxima of the spectra.
    ##For now, one should be careful with features in the spectra.
    E0_diff = np.amax(dydx)
    local_diff = np.argmax(dydx)
    #print(E0_diff, local_diff) #DEBUG
    E0x = x[local_diff]
    E0y = y[local_diff]
    #plt.scatter(E0x,E0y) #DEBUG
    #plt.show() #DEBUG
    
    return E0x, E0y
```

```python
#Função construída por James Moraes de Almeida
def XANES_normalization (df):
    """Recebe um dataset com dados de energia e absorbância e normaliza o dataset."""
    colum = list(df.columns)
    print('Starting Normalization of XANES Spectra')
    
    #Call the E0 calculation function and map it to the input spectra.
    E0x, E0y = XANES_E0(df)
    E0x_index=np.where(df[colum[0]]==E0x)[0]
    print('Spectra starts at',df[colum[0]].iloc[0])
    print('Spectra ends at',df[colum[0]].iloc[-1])
    print('E0 in',E0x_index[0],float(df[colum[0]].iloc[E0x_index].values))
    print('Point interval in background',(float(df[colum[0]].iloc[E0x_index].values)-df[colum[0]].iloc[0])/E0x_index)

    #Define the initial and final points of the linear fit of the background.
    ##This definition seemed good for the the tested spectra, might be changed after more tests are performed.
    start = int(E0x_index/15)
    finish = int(E0x_index/3)
    print('Background Fitting Range',start,finish)
    #Define the background
    background = df[start:finish]
    print('Este eh o valor de background (pre-borda)', background, sep='\t')
    #Linear model for the pre-edge.
    data_x = background.iloc[:, 0].values
    data_y = background.iloc[:, 1].values
    #print("DEBUG", data_x, data_y)
    #Perform the linear fit.
    resultado_fit = linregress(data_x, data_y)
    xwide = df.iloc[:, 0].values
    #print("DEBUG1.5", xwide, resultado_fit)
    #Extrapolate for the whole range.
    #df.plot(x='Energy',y='Absorption',label='Input') #DEBUG
    predicted_initial_range = np.array(resultado_fit.intercept + 
                                     resultado_fit.slope*xwide)
    
    #absorbance = df[colum[1]].values#DEBUG
    #print("DEBUG1.1", predicted_initial_range)
    #Redefine the Absorption column
    df[colum[1]]=df[colum[1]].values - predicted_initial_range
    #print("DEBUG1", df)
    
    #Start the needed variables.
    slope_min = 100000
    np_end = len(df[colum[0]]) #-1 #End of the interval. TODO: Check if we need to discard 
                #a range at the end of the spectra, although as this is for XANES, 
                #I don't think it is needed.
            
    np_start = E0x_index[0]
    
    rmse_min = 1000
    
    #for npt in range(-3, -100, -1):
    for npt in range(np_start,len(df[colum[0]])-5): #, -1):
        #print(npt)
        np_init = npt #Initial point of the interval

        #Define the range for the linear fit.
        final_medida = df.iloc[np_init:np_end] 
        faixa_final = df[np_init:np_end]
        #print("DEBUG2", faixa_final)
        #Defining again the data in range, dumb! TODO: Fix this!
        data_x = faixa_final.iloc[:, 0].values
        data_y = faixa_final.iloc[:, 1].values
        
        #Perform the fit
        resultado_fit = linregress(data_x, data_y)
        predicted_resulto_fit = np.array(resultado_fit.intercept + 
                                     resultado_fit.slope*data_x)
        #print("DEBUG3", data_y, predicted_resulto_fit)
        rmse = mean_squared_error(data_y, predicted_resulto_fit, squared=True)  
        #print('Found Slope of', resultado_fit.slope, 'for', np_init, np_end) #DEBUG
#         print('DEBUG1', npt, rmse, rmse_min)
        #Test if the slope is smaller then the smallest found yet.
        if abs(resultado_fit.slope) < abs(slope_min):
            #print(npt_min, resultado_fit.slope, slope_min, abs(resultado_fit.slope) < abs(slope_min)) #DEBUG
            slope_min = resultado_fit.slope
            min_intercept = resultado_fit.intercept
            npt_min = npt

    #print('Optimal point', npt_min)
    #Defining the range from the lowest slope found.
    final_medida = df.iloc[npt_min:np_end]
    faixa_final = df[npt_min:np_end]
    
    #Defining again the data in range, dumb! TODO: Fix this!
    data_x = faixa_final.iloc[:, 0].values
    data_y = faixa_final.iloc[:, 1].values

    #Perform the linear fit again. TODO: No need, we could just get from the already found parameters, although I did not saved the intercepts.
    resultado_fit_final = linregress(data_x, data_y)

    #Extrapolate for the whole range.
    xwide = df.iloc[:, 0]
    predicted_faixa_final = np.array(resultado_fit_final.intercept + 
                                     resultado_fit_final.slope*xwide)
    #print("Resultado final",resultado_fit_final.intercept, resultado_fit_final.slope)
    #print(absorcao,predicted_faixa_final) #DEBUG
    #print(np.average(predicted_faixa_final)) #DEBUG

    #Normalized spectra.
#     fit_final = df[colum[1]]/np.average(predicted_faixa_final)
    fit_final = df[colum[1]]/(predicted_faixa_final)

    #For visualizations and DEBUG.
    #plt.plot(data_x,resultado_fit.intercept + resultado_fit.slope*data_x, label='Linear Fits', color='b')
    #plt.plot(data_x,min_intercept + slope_min*data_x, label='Best Linear Fit', color='g', linewidth=1.5)
    #plt.plot(df['Energy'].values,df['Absorption'].values,label='Input Spectra',color='black')
    #'''
    plt.plot(df[[colum[0]]].values,fit_final,label='Normalized Spectra', color='red')
    #plt.plot(df[colum[0]].values, df[colum[1]].values, label='Unnormalized Spectra', color='black')
    plt.legend(fontsize=16)

    plt.xlabel("Energy (eV)", fontproperties=font_manager.FontProperties(size=12))
    plt.ylabel("Absorption", fontproperties=font_manager.FontProperties(size=12))
    plt.xticks(fontsize=16)
    plt.yticks(fontsize=16)
    plt.scatter(df[[colum[0]]].iloc[start],fit_final[start], label='Initial Point of bgnd energy')
    plt.scatter(df[[colum[0]]].iloc[finish],fit_final[finish], label='Final Point of bgnd energy')
    plt.scatter(E0x,float(fit_final[E0x_index]), label='E0 value')
    plt.scatter(df[[colum[0]]].iloc[npt_min],fit_final[npt_min], color='b', label='Optimal Point')
    #plt.scatter(npt_min,float(fit_final[npt_min]), label='Optimal point')

    #plt.savefig('plot.png', dpi=600, bbox_inches='tight', transparent=True)
    plt.legend()
    plt.show()
    
    #'''
    return fit_final
```

```python
#Código construído por Gustavo Uchôa e Izaque Silva
def vasculha_arquivos(arquivos_xdi):
    """Essa função lê todos os arquivos xdi e chama as funções de normalização correspondentes.
    
    Argumentos:
        arquivos_xdi: lista contendo o caminho dos arquivos xdi a serem analizados.
    """
    lista_erros = []
    lista_fluorescencia = []
    for arquivo in arquivos_xdi:
        caminho_arquivo = arquivo

        secoes = [
            "Element.symbol",
            "Element.edge",
            "Mono.d_spacing",
            "Mono.name",
            "Sample.formula",
            "Sample.name",
            "Sample.prep",
            "Sample.temperature",
            "Sample.reference",
            "Detector.I0",
            "Detector.I1",
            "Detector.I2",
            "Facility.Name",
            "Beamline.Name",
            "Beamline.name",
            "Facility.name",
            "Beamline.xray_source",
            "Beamline.Storage_Ring_Current",
            "Beamline.I0",
            "Beamline.I1",
            "Scan.start_time",
            "Scan.end_time",
            "ScanParameters.Start",
            "ScanParameters.ScanType",
            "ScanParameters.E0",
            "ScanParameters.Legend",
            "ScanParameters.Region1",
            "ScanParameters.Region2",
            "ScanParameters.Region3",
            "ScanParameters.End"
            ]
        regex = '|'.join(map(re.escape, secoes))

        with open(caminho_arquivo, 'r') as texto:
            linhas = texto.read()
            matches = re.findall(f'({regex}):\\s*(.*)', linhas)
        dicio = {}
        for match in matches:
            secao, valor = match[0], match[1]
            secao_primaria, secao_secundaria = secao.split('.')
            if secao_primaria not in dicio:
                dicio[secao_primaria] = {}
            dicio[secao_primaria][secao_secundaria] = valor

        match = re.search(r'#---+', linhas, re.MULTILINE)
        if match:
            tabela_inicio = match.end()
            tabela_linhas = linhas[tabela_inicio:].strip().split('\n')
            valores_tabela = []
            for linha in tabela_linhas:
                if re.match(r'\s*\d+\.\d+(?:E[+-]?\d+)?\s*', linha):
                    valores_tabela.append([float(valor) for valor in linha.split()])
        else:
            # Se não encontrar o início da tabela, definir valores_tabela como None
            valores_tabela = None

        with open(caminho_arquivo, "r") as arquivo:

            texto = arquivo.read()

        # Encontrar nomes das colunas
        nomes_colunas = re.findall(r"# Column\.\d+: (\w+)", texto)
        # Separar as linhas de dados e criar dicionário associando cada valor ao seu respectivo nome de coluna
        dados = {}
        linhas = texto.splitlines()
        match = re.search(r'#---+', texto, re.MULTILINE)
        for nome in nomes_colunas:
            dados[nome] = []
        if match:
            inicio = match.end()
            tab_linhas = texto[inicio:].strip().split('\n')
            for linha in tab_linhas:
                if re.match(r'\s*\d+\.\d+(?:E[+-]?\d+)?\s*', linha):
                    valores = linha.split()
                    for nome,valor in zip(nomes_colunas,valores):
                        dados[nome].append(float(valor))


        try:
            energy_df = dados["energy"]
        except KeyError:
            energy_df = None
        try:
            i0_df = dados["i0"]
        except KeyError:
            i0_df = None
        try:
            itrans_df = dados["itrans"]
        except KeyError:
            itrans_df = None
        try:
            irefer_df = dados["irefer"]
        except KeyError:
            irefer_df = None
        try:
            ifluor_df = dados["ifluor"]
        except KeyError:
            ifluor_df = None   
        try:
            mutrans_df = dados["mutrans"]
        except KeyError:
            mutrans_df = None
        try:
            mufluor_df = dados["mufluor"]
        except KeyError:
            mufluor_df = None
        

        try:
            #print('Entrei no try, valores de energia', energy_df, sep='\n')
            energy_1 = ",".join(str(element) for element in energy_df)
        except TypeError:
            pass
        try:
            i0_1 = ",".join(str(element) for element in i0_df)
        except TypeError:
            pass
        try:
            itrans_1 = ",".join(str(element) for element in itrans_df)
        except TypeError:
            pass
        try:
            mutrans_1 = ",".join(str(element) for element in mutrans_df)
        except TypeError:
            pass
        try:
            irefer_1 = ",".join(str(element) for element in irefer_df)
        except TypeError:
            pass
        try:
            ifluor_1 = ",".join(str(element) for element in ifluor_df)
        except TypeError:
            pass
        try:
            mufluor_1= ",".join(str(element) for element in mufluor_df)
        except TypeError:
            pass

        def xas_xanes(energia):
            #print(f"O arquivo {arquivo} possui a energia {energia}") #DEBUG
            try:
                maximo = max(energia)
                minimo = min(energia)
                #print(maximo, minimo) #DEBUG


                num = float(maximo) - float(minimo)
                if num > 5 and num < 200:
                    print(f"O arquivo {caminho_arquivo} é XANES!")
                    return 2
                else:
                    return 1
            except:
                lista_erros.append(caminho_arquivo)


        type = xas_xanes(energy_df)

        title = "Populando_a_base"
        if type == 2:

            try:
                element_symbol = dicio["Element"]["symbol"]
            except KeyError:
                element_symbol = "Not Informed"

            try:
                element_edge = dicio["Element"]["edge"]
            except KeyError:
                element_edge = "Not Informed"
            try:
                mono_d_spacing = dicio["Mono"]["d_spacing"]
            except KeyError:
                mono_d_spacing = "Not Informed"
            try:
                mono_name = dicio["Mono"]["name"]
            except KeyError:
                mono_name = "Not Informed"
            try:
                sample_formula = dicio["Sample"]["formula"]
            except KeyError:
                sample_formula = "Not Informed"
            try:
                sample_name = dicio["Sample"]["name"]
            except KeyError:
                sample_name = "Not Informed"
            try:
                sample_prep = dicio["Sample"]["prep"]
            except KeyError:
                sample_prep = "Not Informed"
            try:
                sample_temperature = dicio["Sample"]["temperature"]
            except KeyError:
                sample_temperature = "Not Informed"
            try:
                sample_reference = dicio["Sample"]["reference"]
            except KeyError:
                sample_reference = "Not Informed"
            try:
                detector_I0 = dicio["Detector"]["I0"]
            except KeyError:
                detector_I0 = "Not Informed"
            try:
                detector_I1 = dicio["Detector"]["I1"]
            except KeyError:
                detector_I1 = "Not Informed"
            try:
                detector_I2 = dicio["Detector"]["I2"]
            except KeyError:
                detector_I2 = "Not Informed"
            try:
                facility_Name = dicio["Facility"]["name"]
            except KeyError:
                facility_Name = "Not Informed"
            try:
                beamline_xray_source = dicio["Beamline"]["xray_source"]
            except KeyError:
                beamline_xray_source = "Not Informed"
            try:
                beamline_Storage_Ring_Current = dicio["Beamline"]["Storage_Ring_Current"]
            except KeyError:
                beamline_Storage_Ring_Current = "Not Informed"
            try:
                beamline_I0 = dicio["Beamline"]["I0"]
            except KeyError:
                beamline_I0 = "Not Informed"
            try:
                beamline_I1 = dicio["Beamline"]["I1"]
            except KeyError:
                beamline_I1 = "Not Informed"
            try:
                scan_start_time = dicio["Scan"]["start_time"]
            except KeyError:
                scan_start_time = "Not Informed"
            try:
                scan_end_time = dicio["Scan"]["end_time"]
            except KeyError:
                scan_end_time = "Not Informed"
            try:
                scanParameters_Start = dicio["ScanParameters"]["Start"]
            except KeyError:
                scanParameters_Start = "Not Informed"
            try:
                scanParameters_ScanType = dicio["ScanParameters"]["ScanType"]
            except KeyError:
                scanParameters_ScanType = "Not Informed"
            try:
                scanParameters_E0 = dicio["ScanParameters"]["E0"]
            except KeyError:
                scanParameters_E0 = "Not Informed"
            try:
                scanParameters_Legend = dicio["ScanParameters"]["Legend"]
            except KeyError:
                scanParameters_Legend = "Not Informed"
            try:
                scanParameters_Region1 = dicio["ScanParameters"]["Region1"]
            except KeyError:
                scanParameters_Region1 = "Not Informed"
            try:
                scanParameters_Region2 = dicio["ScanParameters"]["Region2"]
            except KeyError:
                scanParameters_Region2 = "Not Informed"
            try:
                scanParameters_Region3 = dicio["ScanParameters"]["Region3"]
            except KeyError:
                scanParameters_Region3 = "Not Informed"
            try:
                scanParameters_End = dicio["ScanParameters"]["End"]
            except KeyError:
                scanParameters_End = "Not Informed"
            try:
                energy = energy_1
            except KeyError:
                energy = "Not Informed"
            try:
                i0 = i0_1
            except KeyError:
                i0 = "Not Informed"
            try:
                itrans = itrans_1
            except KeyError:
                itrans = "Not Informed"
            try:
                irefer = irefer_1
            except KeyError:
                irefer = "Not Informed"
                #ver o que está causando erro quando descomenta essa parte do código...
#             try:
#                 ifluor = ifluor
#             except KeyError:
#                 ifluor = "Not Informed"
            #try:
                #mufluor = mufluor
            #except KeyError:
                #mufluor = "Not Informed"
            #try:
                #mutrans = mutrans
            #except KeyError:
                #mutrans = "Not Informed"

            df = pd.DataFrame()
            df['energy'] = energy_df
            df['i0'] = i0_df
            df['itrans'] = itrans_df
            df['irefer'] = irefer_df
            df['ifluor'] = ifluor_df
            df['mufluor'] = mufluor_df
            df['mutrans'] = mutrans_df
            #print(ifluor_df, mufluor_df)
            #condições para calcular a absorbância
            if ifluor_df != None and mufluor_df==None:
                print("Fluorescencia")
                df['absorbancia'] = -df['i0']/df['ifluor']
            elif mutrans_df==None:
                print("Trasmissao")
                df['absorbancia'] = np.log(df['i0']/df['itrans'])
                

            xanes = df[['energy', 'absorbancia']]
            
            if ifluor_df != None or mufluor_df != None:
                try:
                    #XANES_normalization_fluorescence(xanes)
                    lista_fluorescencia.append(caminho_arquivo)
                except:
                    print(f"o arquivo {caminho_arquivo} está com erro")
            
            else:
                try:
                    XANES_normalization(xanes)
                except:
                    print(f"o arquivo {caminho_arquivo} está com erro")

    print(f"Os arquivos que possuem erros são: {lista_erros}")
    print(f"Os arquivos espectros de fluorescência são: {lista_fluorescencia}")
    
    #a parte de baixo será utilizada quando o código estiver na base de dados, por enquanto está comentado.
    #connection = sqlite3.connect('db.sqlite3')
    #cursor = connection.cursor()

    # cursor.execute("DELETE FROM database_experiment")
    #cursor.execute('''INSERT INTO database_experiment (experiment_type, experiment_title, xdi_file, additional_info, doi, user_id,element_symbol, element_edge, mono_d_spacing, mono_name, sample_formula, sample_name, sample_prep, sample_temperature, sample_reference, detector_I0, detector_I1, detector_I2, facility_Name, beamline_xray_source, beamline_Storage_Ring_Current, beamline_I0, beamline_I1, scan_start_time, scan_end_time, scanParameters_Start, scanParameters_ScanType, scanParameters_E0, scanParameters_Legend, scanParameters_Region1, scanParameters_Region2, scanParameters_Region3, scanParameters_End, energy,itrans, i0, reference) VALUES (?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?)''',(type,title,f"XDIs/{caminho_arquivo}","Not informed", "Not informed", 2,element_symbol,element_edge,mono_d_spacing,mono_name,sample_formula,sample_name,sample_prep,sample_temperature,sample_reference,detector_I0,detector_I1,detector_I2,facility_Name,beamline_xray_source,beamline_Storage_Ring_Current,beamline_I0,beamline_I1,scan_start_time,scan_end_time,scanParameters_Start,scanParameters_ScanType,scanParameters_E0,scanParameters_Legend,scanParameters_Region1,scanParameters_Region2,scanParameters_Region3,scanParameters_End,energy,i0,itrans,irefer))

    #connection.commit()
    #connection.close()

    # Deletar dados da base  
    # connection = sqlite3.connect('db.sqlite3')
    # cursor = connection.cursor()

    # cursor.execute("DELETE FROM database_experiment")

    # connection.commit()
    # connection.close()
```
